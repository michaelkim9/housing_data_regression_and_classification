### 1. ElasticNet with Lasso
* As there are a lot of features - 247 columns - going to SelectKBest to choose top features and use Lasso method with SelectFromModel.
* Going to train model based on ElasticNet as the regressor.
* Scaling the data with standard scaler so that we could standardize the data set for the model
* Putting all these elements into a pipeline and then tuning hyper parameters in a grid search

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Lasso, Ridge
from sklearn.model_selection import GridSearchCV, ShuffleSplit
from sklearn.feature_selection import SelectKBest, SelectFromModel, f_regression, RFE
from sklearn.linear_model import ElasticNet
```

```python
el_pipe = Pipeline([
    ('skb', SelectKBest(score_func=f_regression)),
    ('scaler', StandardScaler()),
    ('sfm', SelectFromModel(Lasso(random_state=42))),
    ('regr', ElasticNet(random_state=42))
])

el_params = {
    'regr__l1_ratio':[.1,.3,.5,.7,.9],
    'regr__alpha':np.logspace(-1,1,20),
    'skb__k':[140, 160, 180],
    'sfm__estimator':[Lasso(random_state=42),Ridge(random_state=42)]
}

el_gs = GridSearchCV(el_pipe, param_grid=el_params, n_jobs=-1, cv=5)
```

```python
el_gs.fit(X_train, y_train)
```

###### 1. Elastic Net Regressor - Scores

The scores turned out okay. The cross validation scores is just below 80%. The training and test scores were above 80%. Overall the model didn't perform extremeley poorly but going to try another model to see if we can get better scores

```python
el_gs.best_score_
```

```python
el_gs.score(X_train,y_train)
```

```python
el_gs.score(X_test,y_test)
```

###### 1. Elastic Net Regressor - Feature Selection

The SelectKBest chose 140 features which is based on the ANOVA variance. It selects features based on group means and p-values.

The Lasso chose 44 features by restricting coefficients to zero.

And when we look at the coefficients of the features that the Lasso selected and we could get an idea of which features were the greatest predictors of sale price. From the graph below, it looks like "GarageCars" had the greatest influence - which would kind of make sense since the larger home that are valued higher would be the ones that have larger garage sizes. And some of the other features that have large coefficients are ones that deal with square footage.

```python
skb_features = el_gs.best_estimator_.named_steps['skb'].get_support()
sfm_features = el_gs.best_estimator_.named_steps['sfm'].get_support()
print(skb_features.sum())
print(sfm_features.sum())
```

```python
el_feat_importances = el_gs.best_estimator_.named_steps['regr'].coef_
el_cols_and_imps = list(zip(X_train.columns, el_feat_importances))
el_cols_and_imps.sort(key=lambda x:x[1], reverse=False)

el_feat_imp_df = pd.DataFrame(el_cols_and_imps)
el_feat_imp_df.rename(columns={1:'coefficients',0:'column'}, inplace=True)
el_feat_imp_df = el_feat_imp_df

el_feat_imp_df.plot.barh('column','coefficients', figsize=(15,20))
```

###### 1. Elastic Net Regressor - Best Parameters

```python
el_gs.best_params_
```

```python
el_gs_results = pd.DataFrame(el_gs.cv_results_)

el_gs_results.sort_values('rank_test_score', ascending=True).T
```

### 1. RandomForest Regressor with Feature Selection

Going to use SelectKBest in order to select features and then Lasso it to find the best predictors. Then run those through a RandomForestRegressor. Similar to the model above but instead, going to use RandomForest as the regressor.

```python
rf_pipe = Pipeline([
    ('skb', SelectKBest(score_func=f_regression)),
    ('scaler', StandardScaler()),
    ('sfm', SelectFromModel(Lasso(random_state=42))),
    ('reg',RandomForestRegressor(random_state=42))
])

rf_params = {
    'reg__n_estimators':range(90,180,30),
    'reg__max_depth':[1,5,10,12,18,None],
    'skb__k':[140, 160, 180],
    'sfm__estimator':[Lasso(random_state=42),Ridge(random_state=42)]
}

rf_gs = GridSearchCV(rf_pipe, param_grid=rf_params, n_jobs=-1, cv=5)
```

```python
rf_gs.fit(X_train, y_train)
```

###### 1. Random Forest Regressor - Scores

The scores seem to be better for the Random Forest Regressor model. The best score from cross validation was 83% and the training score was very high at 97%. The test score was a little lower at 88% which would indicate that this model may be slightly overfit as it didn't generalize to the test data that well which is indicated by the lower test score. This would be more typical of models dealing with decision trees. But overall, it seems that this model is performing much better than the previous model

###### 1. Random Forest - Best Parameters

With the Random Forest model, the thing to watch out for is the model being overfit. And the way that a Random Forest model could get overfit is if the max depth is too high because the deeper the tree, the more it won't be able to generalize to new data. Looking at the cross valiation scores and the max depth of the tree that was used, it looks like the mean test scores peaks around a max depth of 12. Most of the top models selected a max depth of 18 but the mean test scores generally stayed around the same between max depth of 12 and 18. Originally when I ran this model, I was testing it with higher max depths but noticed it overfitting and so decreased it to what it is now.

## 3. Data Cleaning - Abnormal Sales column

As we are going to be conducting a classification analysis to show whether a sale was normal or abnormal, we need to create a binary column indicating whether a sale was abnormal not. The value counts above show that there are 96 properties classified as abnormal, which we will mark as 1 in a new column we'll create labeled "AbnormalSale". The other property sales will be marked with a 0, making the new "AbnormalSale" column a binary column, which will be our target.

## 3. Boosting Models
* Going to use the **Precision-Recall Curve AUC scoring function** in the GridSearchCV since this is extremely unbalanced data with more than 90% of the data being normal sales. So will use the AUC scoring function
* Going to us `GridSearchCV` to tune hyperparameters for the following ensemble models:
  * `AdaBoostClassifier`
  * `GradientBoostingClassifier`
  * `XGBoostClassifier`
* Conducting analysis of results with the confusion matrix and classification report. From these reports, going to choose the best model to classify abnormal sales.
* It's important to optimize the AUC of the precision/recall curve instead of accuracy because of the class imbalance. The accuracy score will be high sheerly because of imbalance.

```python
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report, auc, make_scorer, precision_recall_curve, roc_curve, roc_auc_score
from sklearn.model_selection import cross_val_score
```

```python
#Generates the Area Under the Curve for precision and recall
def pr_auc_score(y_true, y_score):
    precision, recall, thresholds = precision_recall_curve(y_true, y_score[:, 1])
    return auc(recall, precision, reorder=True)
pr_auc_scorer = make_scorer(pr_auc_score, greater_is_better=True, needs_proba=True)
```

### 3. AdaBoost Model

```python
adaboost_pipe = Pipeline([
    ('scaler',StandardScaler()),
    ('clf',AdaBoostClassifier(DecisionTreeClassifier(max_depth=1,random_state=42),random_state=42))
])

ada_params = {
    'clf__learning_rate':[.1,.5,1,10],
    'clf__n_estimators':[250]
}

ada_gs = GridSearchCV(adaboost_pipe, ada_params, cv=5, n_jobs=-1, scoring=pr_auc_scorer)
```
